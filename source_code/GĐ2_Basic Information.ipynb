```python
df.dtypes
```

Nhận xét:
- Kiểu dữ liệu:
  - Có 1 thuộc tính có kiểu dữ liệu boolean (`bool`).
  - Có 2 thuộc tính có kiểu dữ liệu số thực (`float64`).
  - Có 5 thuộc tính có kiểu dữ liệu số nguyên (`int64`).
  - Có 15 thuộc tính có kiểu dữ liệu chuỗi hoặc `object`.
- Cột `release_date` đang có kiểu dữ liệu là `object`. Để đúng về ý nghĩa, ta sẽ chuyển về dạng `datetime`.

# **2.7. Tình trạng khuyết dữ liệu**

```python
null_df = pd.DataFrame(df.isnull().sum(), columns=["Số mẫu bị thiếu"])
null_df["Tỉ lệ bị thiếu"] = np.round(df.isnull().sum() / df.shape[0] * 100, 5)
null_df.index.name = "Thuộc tính"
null_df = null_df.sort_values(by="Tỉ lệ bị thiếu", ascending=False)
print("Tình trạng khuyết dữ liệu:")
null_df.loc[null_df["Tỉ lệ bị thiếu"] > 0]
```

Nhận xét:
- Thuộc tính `homepage`, `tagline`, `backdrop_path`: Tỉ lệ thiếu rất cao, có thể ảnh hưởng đến khả năng phân tích dữ liệu về trang chủ của bộ phim.
- Thuộc tính `production_companies`, `imdb_id`, `production_countries`, `spoken_languages`, `genres`, `poster_path`: Tỉ lệ thiếu khá cao, thông tin về các thông tin này có thể không đầy đủ.
- Thuộc tính `overview`, `release_date`: Tỉ lệ thiếu không quá cao.

# **2.8. Với mỗi cột số, các giá trị được phân bố như thế nào?**

```python
numeric_df = df.select_dtypes(include=["number"])
nume_col_df = numeric_df.describe()
nume_col_df.loc["missing_count"] = numeric_df.isna().sum()
nume_col_df.loc["missing_ratio"] = (numeric_df.isna().sum() / numeric_df.shape[0]) * 100
nume_col_df
```

# **2.9. Với mỗi cột phân loại, các giá trị được phân bổ như thế nào?**

```python
object_df = df.select_dtypes(include=["object", "bool"])
object_col_df = object_df.describe()
object_col_df.loc["missing_count"] = object_df.isna().sum()
object_col_df.loc["missing_ratio"] = (object_df.isna().sum() / object_df.shape[0]) * 100
object_col_df
```
